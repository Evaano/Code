```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

def cluster_and_visualise(datafilename, K, featurenames):
    data = np.genfromtxt(datafilename, delimiter=',')
    transposed_data = data.T

    fig, axs = plt.subplots(len(transposed_data), len(transposed_data))
    fig.set_size_inches(28, 28)

    kmeans = KMeans(n_clusters=K)
    clusters = kmeans.fit_predict(transposed_data.T)

    cluster_colors = ['blue', 'green', 'orange']
    cmap = plt.cm.get_cmap('plasma', K)

    for i, row_axs in enumerate(axs):
        for j, ax in enumerate(row_axs):
            if i == j:
                hist_data = transposed_data[i]
                ax.hist(hist_data[:len(hist_data)//2], bins=10, color=cluster_colors[0], alpha=0.5, label='Cluster 1')
                ax.hist(hist_data[len(hist_data)//2:], bins=10, color=cluster_colors[1], alpha=0.5, label='Cluster 2')
                ax.set_xlabel('')
                ax.set_ylabel('')
                ax.legend()
            else:
                for cluster_label in range(K):
                    cluster_data = data[clusters == cluster_label].T 
                    ax.scatter(cluster_data[j], cluster_data[i], c=cmap(cluster_label), marker='o', alpha=0.7)
                    ax.set_xlabel('')
                    ax.set_ylabel('')

            if i == len(transposed_data) - 1:
                ax.set_xlabel(featurenames[j])
            if j == 0:
                ax.set_ylabel(featurenames[i])

            ax.grid()


    fig.suptitle(f"Thaaib, K={K}")
    plt.savefig("myVisualisation.png")
    plt.show()
    return fig, axs

if __name__ == "__main__":
    featurenames = ('mean_red', 'mean_green', 'mean_blue', 'width', 'height', 'weight')
    datafile = 'fruit_values.csv'
    cluster_and_visualise(datafile, 3, featurenames)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

def cluster_and_visualise(datafilename, K, featurenames):
    data = np.genfromtxt(datafilename, delimiter=',')
    transposed_data = data.T

    fig, axs = plt.subplots(len(transposed_data), len(transposed_data))
    fig.set_size_inches(28, 28)

    kmeans = KMeans(n_clusters=K)
    clusters = kmeans.fit_predict(transposed_data.T)
    
    # Generate cluster colors based on the number of clusters (K)
    cluster_colors = plt.cm.jet(np.linspace(0, 1, K))

    for i, row_axs in enumerate(axs):
        for j, ax in enumerate(row_axs):
            if i == j:
                for cluster_label in range(K):
                    hist_data = transposed_data[i][clusters == cluster_label]
                    ax.hist(hist_data, bins=10, color=cluster_colors[cluster_label], alpha=0.5, label=f'Cluster {cluster_label + 1}')

                ax.set_xlabel('')
                ax.set_ylabel('')
                ax.legend()
            else:
                for cluster_label in range(K):
                    cluster_data = data[clusters == cluster_label].T 
                    ax.scatter(cluster_data[j], cluster_data[i], c=cluster_colors[cluster_label], marker='o', alpha=0.7)
                    ax.set_xlabel('')
                    ax.set_ylabel('')

            if i == len(transposed_data) - 1:
                ax.set_xlabel(featurenames[j])
            if j == 0:
                ax.set_ylabel(featurenames[i])

            ax.grid()

    fig.suptitle(f"Hassan Shaadhu Mohamed Hassan, K={K}")
    plt.savefig("myVisualisation.png")
    plt.show()
    return fig, axs

if __name__ == "__main__":
    featurenames = ('mean_red', 'mean_green', 'mean_blue', 'width', 'height', 'weight')
    datafile = 'fruit_values.csv'
    cluster_and_visualise(datafile, 3, featurenames)
```
